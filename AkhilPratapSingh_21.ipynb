```python
#AkhilPratapSingh_21

import numpy as np

def check_winner(board):
    """Checks if there's a winner."""
    for row in board:
        if all(cell == 'X' for cell in row):
            return 'X'
        if all(cell == 'O' for cell in row):
            return 'O'

    for col in range(3):
        if all(board[row][col] == 'X' for row in range(3)):
            return 'X'
        if all(board[row][col] == 'O' for row in range(3)):
            return 'O'

    if all(board[i][i] == 'X' for i in range(3)) or all(board[i][2-i] == 'X' for i in range(3)):
        return 'X'
    if all(board[i][i] == 'O' for i in range(3)) or all(board[i][2-i] == 'O' for i in range(3)):
        return 'O'

    return None

def find_best_move(board):
    """Finds the next best move (first empty cell)."""
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                return (i, j)
    return None

def print_board(board):
    """Prints the Tic-Tac-Toe board."""
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def main():
    """Runs the Tic-Tac-Toe solver."""
    board = [[' ' for _ in range(3)] for _ in range(3)]

    while True:
        print_board(board)
        winner = check_winner(board)
        if winner:
            print(f"{winner} wins!")
            break

        move = find_best_move(board)
        if not move:
            print("It's a tie!")
            break

        row, col = move
        board[row][col] = 'X'  # Assuming 'X' is the AI/player
        print(f"AI plays at ({row}, {col})\n")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the dataset
tic_tac_toe_data = {
    "board_state": [
        "XOX,O X, XO", "XX ,OO ,   ", "XOX,OXO,X  ", "XO ,OX ,X O",
        "XOX,XOO,OX ", "XXO,OXX,OO ", "XO ,OX ,XOX", "XOX,OXO,XOX"
    ],
    "best_move": ["(1,1)", "(0,2)", "(2,2)", "(1,2)", "(2,2)", "(2,2)", "(1,2)", "None"],
    "winner": ["None", "X", "X", "None", "O", "O", "None", "Tie"]
}

# Convert to DataFrame
tic_tac_toe_df = pd.DataFrame(tic_tac_toe_data)

# Plot 1: Winner Distribution
winner_counts = tic_tac_toe_df["winner"].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(winner_counts.index, winner_counts.values, color=['blue', 'red', 'green', 'gray'])
plt.xlabel("Winner")
plt.ylabel("Count")
plt.title("Distribution of Winners in Tic-Tac-Toe Dataset")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot 2: Best Move Frequency (excluding 'None' moves)
best_move_counts = tic_tac_toe_df[tic_tac_toe_df["best_move"] != "None"]["best_move"].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(best_move_counts.index, best_move_counts.values, color='purple', edgecolor='black')
plt.xlabel("Best Move (Row, Col)")
plt.ylabel("Frequency")
plt.title("Best Move Frequency in Tic-Tac-Toe Dataset")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
